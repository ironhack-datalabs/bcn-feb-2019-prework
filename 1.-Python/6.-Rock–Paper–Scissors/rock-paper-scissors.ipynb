#### 2. Create a list that includes the 3 possible gesture options of the game: 'rock', 'paper' or 'scissors'. Store the list in a variable called `gestures`.

```python
gestures = ['rock', 'paper', 'scissors']
```

#### 3. Create a variable called `n_rounds` to store the maximum number of rounds to play in a game.

```python
rounds_to_win = 0
```

#### 4. Create a variable called `rounds_to_win` to store the number of rounds that a player must win to win the game.
**Hint**: the value stored in `rounds_to_win` depends on the value of `n_rounds`. 

```python
def rounds():
    rounds_to_win = 0
    n_rounds = int(input("Enter the desired number of rounds. Remember that rounds must be odd: 1, 3, 5, ...  : "))
    while (n_rounds % 2) == 0: 
        print('ERROR! NUMBER OF ROUNDS MUST BE ODD: 1, 3, 5, ..')
        n_rounds = int(input("Enter the desired number of rounds. Remember that rounds must be odd: 1, 3, 5, ...  : "))
    if n_rounds == 1:
        rounds_to_win = 1
    elif n_rounds == 3:
        rounds_to_win = 2
    elif n_rounds > 3 and (n_rounds % 2) != 0:
        rounds_to_win = n_rounds - 2
    return n_rounds, rounds_to_win
```

```python
rounds()
```

#### 5. Define a function that randomly returns one of the 3 gesture options.
You will use this function to simulate the gesture choice of the computer. 

```python
def cpu_gesture():
    cpu_action = rd.choice(gestures)
    return cpu_action
```

#### 6. Define a function that asks the player which is the gesture he or she wants to show: 'rock', 'paper' or 'scissors'.
The player should only be allowed to choose one of the 3 gesture options. If the player's choice is not rock, paper or scissors, keep asking until it is.

```python
def player_gesture():
    player_action = input("Enter a choice: rock, paper, scissors: ").lower()
    while player_action not in gestures:
        player_action = input("Enter a choice: rock, paper, scissors: ").lower()
    return player_action
```

#### 7. Create two variables to store the number of rounds that the computer and the player have won. Call these variables `cpu_score` and `player_score`.

```python
cpu_score = 0
player_score = 0
```

#### 8. Define a function that checks who won a round. 
The function should return 0 if there is a tie, 1 if the computer wins and 2 if the player wins.

```python
def check_winner(cpu_choice, player_choice):
    result = 0
    if player_choice == cpu_choice:
        result = 0
    elif player_choice == "rock":
        if cpu_choice == "scissors":
            result = 2
        else:
            result = 1
    elif player_choice == "paper":
        if cpu_choice == "rock":
            result = 2
        else:
            result = 1
    elif player_choice == "scissors":
        if cpu_choice == "paper":
            result = 2
        else:
            result = 1
    return cpu_choice, player_choice, result
```

```python
cpu_choice = cpu_gesture()
player_choice = player_gesture()

round_winner = check_winner(cpu_choice, player_choice)
```

#### 9. Define a function that prints the choice of the computer, the choice of the player and a message that announces who won the current round. 
You should also use this function to update the variables that count the number of rounds that the computer and the player have won. The score of the winner increases by one point. If there is a tie, the score does not increase.

```python
def round_summary(cpu_choice, user_choice, result):
    cpu_score = 0
    player_score = 0
    
    print('The computer selected: ', cpu_choice)
    print('The user picked: ', user_choice)
    
    if result == 0:
        print("It's a tie")
    elif result == 1:
        print("CPU wins!")
        cpu_score = 1
    elif result == 2:
        print('Player wins!')
        player_score = 1
    return cpu_score, player_score
```

```python
cpu_choice = round_winner[0]
player_choice = round_winner[1]
result = round_winner[2]

round_score = round_summary(cpu_choice, player_choice, result)
```

#### 10. Now it's time to code the execution of the game using the functions and variables you defined above. 

First, create a loop structure that repeats while no player reaches the minimum score necessary to win and the number of rounds is less than the maximum number of rounds to play in a game.  

Inside the loop, use the functions and variables above to create the execution of a round: ask for the player's choice, generate the random choice of the computer, show the round results, update the scores, etc. 

```python
rounds_played = 0
cpu_score = 0
player_score = 0
def_rounds = rounds()
n_rounds = def_rounds[0]
rounds_to_win = def_rounds[1]
#---------------------------------------------------------------------------------------------------------------------
while ((cpu_score < rounds_to_win)  
    and (player_score < rounds_to_win) 
    and (rounds_played < n_rounds)):
    print("Welcome to round " + str(rounds_played + 1))
#---------------------------------------------------------------------------------------------------------------------
    cpu_choice = cpu_gesture()
    player_choice = player_gesture()
    round_winner = check_winner(cpu_choice, player_choice)
#---------------------------------------------------------------------------------------------------------------------
    cpu_choice = round_winner[0]
    player_choice = round_winner[1]
    result = round_winner[2]
    round_score = round_summary(cpu_choice, player_choice, result)
#---------------------------------------------------------------------------------------------------------------------
    cpu_score = round_score[0]
    player_score = round_score[1]
#---------------------------------------------------------------------------------------------------------------------
    print("You have played " + str(rounds_played + 1) + " out of " + str(n_rounds)+" rounds.")
    print("The current score is:\n"+"Computer: " + str(cpu_score)+"\nUser: " + str(player_score))
    rounds_played = rounds_played + 1
```
