```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list
import random
print ('Lets play Rock, Paper and Scissors')
print ('Can you defeat me?')
# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.
options = ['stone', 'paper', 'scissors']

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...
max_games = input("Introduce the number of rounds you want to play:")
max_games = int(max_games)

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games
games_to_win = int(max_games / 2) + 1
print ('The first to win', games_to_win, 'rounds is the winner.')

# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.

def computer_play():
  return random.choice(options)


# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.
def player_play():
  print("""
  Please, introduce your option:
  1 - Stone
  2 - Paper
  3 - Scissors
  """)
  
  good_input = False
  while good_input == False: #Pedimos opcion y checkeamos que el usuario introduzca una de las opciones correctas:
    play = input("Your option: ")
    if play == 'stone' or play == 'paper' or play == 'scissors' or play == '1' or play == '2' or play == '3' or play == 'Stone' or play == 'Paper' or play == 'Scissors':
      good_input = True
    else:
      print("""
      Wrong option.
      Please, introduce one of the available options:
      1 - Stone
      2 - Paper
      3 - Scissors
      """)
    #Convertimos las diferentes opciones en las "estandar"      
  if play == 'Stone' or play == '1':
    play = 'stone'
  elif play == 'Paper' or play == '2':
    play = 'paper'
  elif play == 'Scissors' or play == '3':
    play = 'scissors'     
    
  return play #Devolvemos la jugada escogida

# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins
def combat(computer, player):
  if computer == player:
    return 0
  elif computer == 'stone':
    if player == 'scissors':
      return 1
    else:
      return 2
  elif computer == 'paper':
    if player == 'stone':
      return 1
    else:
      return 2
  else:
    if player == 'paper':
      return 1
    else:
      return 2

# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated
def round(computer_wins, player_wins):
  computer = computer_play()
  player = player_play()
  result = combat(computer, player)
  print ('Computer plays', computer, 'and player plays', player, '.')
  if result ==0:
    print ('Is a draw!')
  elif result==1:
    computer_wins += 1
    print ('Computer wins this round!')
  else:
    player_wins += 1
    print ('Player wins this round!')
    
  print ('The game goes: Computer:', computer_wins, 'and Player:', player_wins)
  return computer_wins, player_wins  


# Create two variables that accumulate the wins of each participant
results = [0,0] #results[0] corresponde a computer y results[1] a player

# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.
for rounds in range (0, max_games):
  print ('Round:', rounds+1)
  results = round(results[0], results[1])
  if max(results)==games_to_win:
    break
    
# Print by console the winner of the game based on who has more accumulated wins
print ('The final result is:')
print ('Computer:', results[0], 'rounds.')
print ('Player:', results[1], 'rounds.')
if results[0] == games_to_win or results[0] > results[1]:
  print ('Computer wins the game!')
elif results[1] == games_to_win or results[1] > results[0]:
  print ('Player wins the game!')
else:
  print ('Ohh, this is a draw. You want to try again?')
```
