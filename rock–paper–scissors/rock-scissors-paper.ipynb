```python
import random
import math
gestures = ['rock', 'paper', 'scissors']
n_rounds = int(input("Number of rounds. Choice an odd number: "))
rounds_to_win = math.floor((n_rounds/2)+1)
cpu_score = 0
player_score = 0

def cpu_turn():
    cpu_chooses = random.choice(gestures)
    print("Cpu choice: ", cpu_chooses)
    print("-------------------------")
    return cpu_chooses

def player_turn():
    player_chooses = 'null'
    while player_chooses not in gestures:
        player_chooses = str(input("Choose between: rock, paper or scissors: "))
    print("-------------------------")
    print("Player choice: ", player_chooses)
    return player_chooses

def winner(Cchoice, Pchoice):
    if Cchoice == Pchoice:
        value = 0
        print("The round was a tie")
        return value
    elif Cchoice == 'rock':
        if Pchoice == 'paper':
            value = 2
            print("The PLAYER win the round")
            return value
        elif Cchoice == 'scissors':
            value = 1
            print("The CPU win the round")
            return value
    elif Cchoice == 'scissors':
        if Pchoice == 'rock':
            value = 2
            print("The PLAYER win the round")
            return value
        elif Pchoice == 'paper':
            value = 1
            print("The CPU win the round")
            return value     
    elif Cchoice == 'paper':
        if Pchoice == 'scissors':
            value = 2
            print("The PLAYER win the round")
            return value
        elif Pchoice == 'rock':
            value = 1
            print("The CPU win the round")
            return value
        
def informer(value):
    if value == 1:
        global cpu_score
        cpu_score +=1
    elif value == 2:
        global player_score
        player_score += 1
        
played_rounds = 0
print("player score initial", player_score)
print("rounds to win", rounds_to_win)
while ((player_score < rounds_to_win) & (cpu_score < rounds_to_win)) & (played_rounds < n_rounds):
    player = player_turn()
    cpu = cpu_turn()
    W = winner(cpu, player)
    informer(W)
    played_rounds += 1
    print("No. Rounds played: ",played_rounds)
    print("Cpu score: ", cpu_score)
    print("Player score: ", player_score)

    print("-------------------------")
if cpu_score > player_score:
    print("CPU wins the match"'\n'"Total wins: ", cpu_score)
elif cpu_score < player_score:
    print("The PLAYER wins the match"'\n'"Total wins: ", player_score)
else:
    print("The match was a tie")
```
